```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.interpolate import interp1d
import scipy.integrate as integrate

import seaborn as sns
from matplotlib.ticker import FormatStrFormatter
```

```python
font = {'family': 'serif', 'size': 16}
plt.rc('font', **font)
plt.rc('text', usetex=True)
```

```python
kb = 3.1668105e-06 # in hartree
ha2ev = 27.211386
kbev = kb*ha2ev # in eV
bohr2a = 0.52917721
```

```python
from matplotlib import cm
colors = cm.brg(np.linspace(0, 1, 20))
colors2 = cm.viridis(np.linspace(0, 1, 10))
```

```python
plist = [50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325,350,375,400]
```

# V-T relations at different P

```python
vtcooling={}
vtcooling={ s: {x: np.asmatrix(np.genfromtxt('./output/enthalpy-volume-curves-'+str(s)+'/V-cooling-P-'+format(x)+'.his'))
         for x in plist} for s in range(1,9)}

petcooling={}
petcooling={ s: {x: np.asmatrix(np.genfromtxt('./output/enthalpy-volume-curves-'+str(s)+'/cooling-P-'+format(x)+'.his'))
         for x in plist} for s in range(1,9)}

vtheating={}
vtheating={ s: {x: np.asmatrix(np.genfromtxt('./output/enthalpy-volume-curves-'+str(s)+'/V-heating-P-'+format(x)+'.his'))
         for x in plist} for s in range(1,9)}

petheating={}
petheating={ s: {x: np.asmatrix(np.genfromtxt('./output/enthalpy-volume-curves-'+str(s)+'/heating-P-'+format(x)+'.his'))
         for x in plist} for s in range(1,9)}
```

```python
fout = 'si-cooling-heating-nst-1728-V.png'
linestyle='solid'
lw=0.3
with sns.plotting_context("notebook", font_scale=1.5):
    with sns.axes_style("ticks",{"xtick.major.size": 8, "ytick.major.size": 8,"xtick.minor.size": 4, "ytick.minor.size": 4}):
        f, ax = plt.subplots(1,  figsize=np.array([0.7,1]) * 6)
 
        for s in [1]:
            for i,t in enumerate(plist):
                ax.plot(vtcooling[s][t][:,0], vtcooling[s][t][:,1]/1728, color=colors[i],label=str(t),linestyle=linestyle, linewidth=1,alpha=0.8)
                ax.plot(vtheating[s][t][:,0], vtheating[s][t][:,1]/1728, color=colors[i],linestyle=linestyle, linewidth=lw,alpha=0.8)
        for s in range(2,9):
            for i,t in enumerate(plist):
                ax.plot(vtcooling[s][t][:,0], vtcooling[s][t][:,1]/1728, color=colors[i],linestyle=linestyle, linewidth=lw,alpha=0.8)
                ax.plot(vtheating[s][t][:,0], vtheating[s][t][:,1]/1728, color=colors[i],linestyle=linestyle, linewidth=lw,alpha=0.8)

        ax.legend(bbox_to_anchor=(1.05, 1.05), loc='upper left', borderaxespad=0.)
        ax.set_xlim([200, 1100]) 
        ax.set_ylim([2000./1728,3.3])
        ax.xaxis.set_major_formatter(FormatStrFormatter('%d'))
        ax.set(xlabel='Temperature [K]', ylabel='Volume per atom [$\AA^3$]')    

        if fout:
            f.savefig(fout, dpi=300, bbox_inches='tight')
plt.show()
```

```python
fout = 'si-cooling-heating-nst-1728-PE.png'
#fout = None
linestyle='solid'
lw=0.3
with sns.plotting_context("notebook", font_scale=1.5):
    with sns.axes_style("ticks",{"xtick.major.size": 8, "ytick.major.size": 8,"xtick.minor.size": 4, "ytick.minor.size": 4}):
        f, ax = plt.subplots(1,  figsize=np.array([0.7,1]) * 6)
        
        for s in range(1,9):
            for i,t in enumerate(plist):
                ax.plot(petcooling[s][t][:,0], petcooling[s][t][:,1]/1728, color=colors[i],label=str(t),linestyle=linestyle, linewidth=lw,alpha=0.8)
                ax.plot(petheating[s][t][:,0], petheating[s][t][:,1]/1728, color=colors[i],linestyle=linestyle, linewidth=lw,alpha=0.8)
            
        #ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
        ax.set_xlim([200, 1100]) 
        #ax.set_ylim([2000.,8000])
        ax.xaxis.set_major_formatter(FormatStrFormatter('%d'))
        ax.set(xlabel='Temperature [K]', ylabel='Potential energy per atom [eV]')    

        if fout:
            f.savefig(fout, dpi=300, bbox_inches='tight')
plt.show()
```
